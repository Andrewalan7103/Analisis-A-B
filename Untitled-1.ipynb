**Parte 2. Análisis del test A/B**

Realizaste una prueba A/B y obtuviste los resultados descritos en los archivos df_completo_us.csv y visits_us.csv.
Ejercicio
Analiza el test A/B:

### *Hipótesis nula (H₀) p_value ≥ alpha :*
No existe una diferencia significativa en las métricas clave (ingresos, tamaño de pedido promedio, tasa de conversión) entre los grupos A y B.


### *Hipótesis alternativa (H₁) p_value < alpha:*
Existe una diferencia significativa en las métricas clave entre los grupos A y B.

Analisis y depuracion de datos.

```python
# Cargar los datos de pedidos y visitas
df_orders = pd.read_csv('orders_us.csv', sep=',')
df_orders['date'] = pd.to_datetime(df_orders['date'], format='%Y-%m-%d')
df_visits = pd.read_csv('visits_us.csv', sep=',')
df_visits['date'] = pd.to_datetime(df_visits['date'], format='%Y-%m-%d')

df_orders.info()
df_visits.info()

# Revisar si existen usuarios duplicados en ambos grupos
# Un usuario debería estar solo en un grupo A o B, no en ambos
df_completo = (
    df_orders
    # 1. Filtrar usuarios duplicados
    .loc[~df_orders["visitorId"].isin(
        df_orders.groupby("visitorId")["group"]
        .nunique()
        .reset_index()
        .query("group > 1")["visitorId"]
    )]
    # 2. Resetear índices
    .reset_index(drop=True)
    # 3. Unir con visitas
    .merge(df_visits, on=["date", "group"], how="right")
)

print(f"Pedidos originales: {df_orders.shape[0]}")
print(f"Pedidos después de depuración: {df_completo.shape[0]}")
df_completo.info()
```

### 9. Encuentra la significancia estadística de la diferencia en la conversión entre los grupos utilizando los datos en bruto. Haz conclusiones y conjeturas.

```python
# Suma total de pedidos y visitas por grupo
conversiones = conversion.groupby('group')['orders'].sum()
visitas = conversion.groupby('group')['visits'].sum()

# Tasa de conversión general por grupo
conversion_rate_A = conversiones['A'] / visitas['A']
conversion_rate_B = conversiones['B'] / visitas['B']

print(f"Tasa de conversión Grupo A: {conversion_rate_A:.4f}")
print(f"Tasa de conversión Grupo B: {conversion_rate_B:.4f}")
print(f"Diferencia absoluta: {conversion_rate_B - conversion_rate_A:.4f}")
print(f"Diferencia relativa: {(conversion_rate_B/conversion_rate_A - 1)*100:.2f}%")

from statsmodels.stats.proportion import proportions_ztest

count = np.array([conversiones['A'], conversiones['B']])
nobs = np.array([visitas['A'], visitas['B']])
stat, pval = proportions_ztest(count, nobs)

print(f"p-value: {pval:.4f}")
if pval < 0.05:
    print("No Existe diferencia significativa en la tasa de conversión.")
else:
    print("Existe diferencia significativa en la tasa de conversión.")
    
    # Tasa de conversión general por grupo (ya calculada)
conversion_rates = [conversion_rate_A, conversion_rate_B]
labels = ['Grupo A', 'Grupo B']

plt.figure(figsize=(6,4))
plt.bar(labels, conversion_rates, color=['blue', 'orange'])
plt.ylabel('Tasa de conversión')
plt.title('Tasa de conversión general por grupo')
plt.ylim(0, max(conversion_rates)*1.2)
for i, v in enumerate(conversion_rates):
    plt.text(i, v + 0.0005, f"{v:.3%}", ha='center', fontweight='bold')
plt.show()
```

De acuerdo con la prueba proportions_ztest y nuestra hipotesis "No existe diferencia significativa en la tasa de conversion" por o tanto la prueba B podemos decir que no esta teniendo un impacto  la conversión de nuevos clientes
